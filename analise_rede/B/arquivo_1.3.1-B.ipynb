```python
import re
import matplotlib.pyplot as plt

# Caminho para o arquivo de experimento escolhido
arquivo = 'exp-005-B-C.txt'

# Lista para armazenar os tempos extraídos
tempos_ms = []

# Abrir e ler o arquivo linha por linha
with open(arquivo, 'r') as f:
    for linha in f:
        # Usar regex para extrair o tempo 
        match = re.search(r'time=([\d.]+) ms', linha)
        if match:
            tempos_ms.append(float(match.group(1)))  # Armazena como número

# Criar eixo X (de 1 a N amostras)
amostras = list(range(1, len(tempos_ms) + 1))

# Plotar o gráfico
plt.figure(figsize=(10, 5))
plt.plot(amostras, tempos_ms, marker='o', linestyle='-', color='blue')
plt.title('Série Temporal de Atraso - exp-005-B-C.txt')
plt.xlabel('Número da Amostra (ping)')
plt.ylabel('Tempo de Resposta (ms)')
plt.grid(True)
plt.tight_layout()
plt.show()
```
